```python
import pandas as pd
#Fuente de informacion https://www.datosabiertos.gob.pe/dataset/resultados-por-mesa-de-las-elecciones-congresales-2021-oficina-nacional-de-procesos-1
file_path = "EG2021_CONGRESAL.csv"

# Cargamos todo el archivo de una vez
df = pd.read_csv(
    file_path,
    sep=";",
    encoding="latin-1"  # si falla prueba con "cp1252"
)
# DESHACER CONTINENTES
extra=    ["EUROPA", "ASIA", "AFRICA", "AMERICA", "OCEANIA"]
# Filtrar voto en el extranjero
df["DEPARTAMENTO"] = df["DEPARTAMENTO"].replace(extra, "EXTRANJERO")

df = df.drop(["TIPO_ELECCION", "TIPO_OBSERVACION", "UBIGEO", "CODIGO_OP", "DESCRIP_ESTADO_ACTA"], axis=1)

# Crear diccionario con un DataFrame por departamento
chunks = {dep: subdf for dep, subdf in df.groupby("DEPARTAMENTO")}

print("Departamentos cargados:", len(chunks))
for dep, subdf in chunks.items():
    print(dep, "→", len(subdf), "filas")
```

```python
chunks["LIMA METROPOLITANA"] = chunks["LIMA"] [df['PROVINCIA'] == 'LIMA']
chunks["LIMA PROVINCIAS"] = chunks["LIMA"] [df['PROVINCIA'] != 'LIMA']
chunks.pop("LIMA")
```

```python
def ExtraerVotos(chunk=ejemplo):
    c=chunk.fillna(0)
    votos=c.groupby("DESCRIPCION_OP")["N_TOTAL_VOTOS"].sum()
    return votos

for i in chunks.keys():
    print(len(ExtraerVotos(chunks[i])))
```

```python
def TabularVotos():
    dic={}
    for i in chunks.keys():
        v= ExtraerVotos(chunks[i])
        dic[i]=v
    res = pd.concat(dic, axis=1).reset_index()
    res = res.rename(columns={'index': 'DESCRIPCION_OP'}).fillna(0)
    return res
tabla=TabularVotos()
tabla
```
